##**Import**

```python
import cv2 # OpenCV
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
cv2.__version__
```

```python
imagem = cv2.imread('/content/Lady_Gaga.jpg')
```

```python
#tamanho da imagem
imagem.shape
```

```python
#Converter a imagem para tons de cinza
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
```

```python
#Tamanho da imagem em tons de cinza
imagem_cinza.shape
```

```python
cv2_imshow(imagem_cinza)
```

```python
detector_facial = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
deteccoes = detector_facial.detectMultiScale(imagem_cinza)
deteccoes
```

```python
for (x, y, w, h) in deteccoes:
  print(x, y, w, h)
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0, 255, 255), 4)
cv2_imshow(imagem)
```

```python
imagem = cv2.imread('/content/Lady_Gaga.jpg')
imagem = cv2.resize(imagem, (0,0), fx = 0.7, fy = 0.7)
imagem.shape
cv2_imshow(imagem)
```

```python
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
deteccoes = detector_facial.detectMultiScale(imagem_cinza)
for (x, y, w, h) in deteccoes:
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0,255,255), 3)
cv2_imshow(imagem)
```

```python
imagem = cv2.imread('/content/Lady_Gaga.jpg')
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
print(imagem_cinza.shape)
deteccoes = detector_facial.detectMultiScale(imagem_cinza, scaleFactor = 1.2)
for (x, y, w, h) in deteccoes:
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0,0,255), 2)
cv2_imshow(imagem)
```

```python
imagem = cv2.imread('/content/seleo_2022.jpg')
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
deteccoes = detector_facial.detectMultiScale(imagem_cinza, scaleFactor = 1.15, minNeighbors = 5)
for (x, y, w, h) in deteccoes:
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0,0,255), 2)
cv2_imshow(imagem)
```

```python
imagem = cv2.imread('/content/seleo_2022.jpg')
imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
deteccoes = detector_facial.detectMultiScale(imagem_cinza, maxSize = (50,50))
for (x, y, w, h) in deteccoes:
  print(w, h)
  cv2.rectangle(imagem, (x, y), (x + w, y + h), (0,255,255), 2)
cv2_imshow(imagem)
```
